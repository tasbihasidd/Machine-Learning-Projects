```python
import pandas as pd
```

```python
df = pd.read_csv('homeprices.csv')
```

```python
df
```

### Now we need to convert only one col in label encodig but using pandas this time

```python
dummies = pd.get_dummies(df.town)
dummies
```

### Merging dummies df and our original df

```python
merged = pd.concat([df,dummies],axis=1)
merged
```

```python
final = merged.drop(['town'], axis='columns')
final
```

### Dummy Variable Trap
When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.

N:
E: sklearn library takes care of dummy variable trap hence even if you don't drop one of the state columns it is going to work, however we should make a habit of taking care of dummy variable trap ourselves just in case library that you are using is not handling this for you

```python
final = final.drop(['west windsor'],axis=1)
final
```

```python
X = final.drop('price',axis=1)
X
```

```python
y = final.price
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict([[2600,True,False]])
```
