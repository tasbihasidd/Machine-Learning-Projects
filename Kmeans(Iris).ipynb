```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = load_iris()
df
```

```python
df2 = pd.DataFrame(df.data,columns=df.feature_names)
df2
```

```python
df2.head()
```

```python
df2['flower']=df.target
df2
```

```python
df2.drop(['sepal length (cm)', 'sepal width (cm)', 'flower'],axis='columns',inplace=True)
df2.head()
```

```python
km = KMeans(n_clusters=3)
y_predict = km.fit_predict(df2)
y_predict
```

```python
df2['clusters'] = y_predict
df2.head()
```

```python
km.cluster_centers_
```

```python
df3 = df2[df2.clusters==0]
df4 = df2[df2.clusters==1]
df5 = df2[df2.clusters==2]
```

```python
plt.scatter(df3['petal length (cm)'],df3['petal width (cm)'],color='blue')
plt.scatter(df4['petal length (cm)'],df4['petal width (cm)'],color='green')
plt.scatter(df5['petal length (cm)'],df5['petal width (cm)'],color='yellow')
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df)
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
